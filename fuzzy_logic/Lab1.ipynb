```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 11 , 0.1)
a = 2
b = 4
c = 6
tr_func = fuzz.trimf(x, [a, b, c])


plt.plot(x , tr_func, 'b' , linewidth =2, label ='Triangular Membership Function')
plt.title('Triangular Fuzzy Set')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Value')
plt.legend()
plt.show()
```

```python
a = 2
b = 4
c = 6
d = 8
trap_func = fuzz.trapmf(x, [a, b, c, d])

plt.plot(x , trap_func, 'b' , linewidth =2, label ='Trapezoidal Membership Function')
plt.title('Trapezoidal Fuzzy Set')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Value')
plt.legend()
plt.show()
```

```python
mean = 6
std_dev = 1
membership_function = fuzz.gaussmf(x, mean, std_dev)

plt.plot(x , membership_function, 'b' , linewidth =2, label ='Gaussian Membership Function')
plt.title('Gaussian Fuzzy Set')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Value')
plt.legend()
plt.show()
```

```python
t_cold = np.arange(10, 21, 1)
t_warm = np.arange(15, 31, 1)

x = np.linspace(5, 35, 100)
tr_func_cold = fuzz.trimf(x, [10, 15, 30])
tr_func_warm = fuzz.trimf(x, [15, 20, 30])

plt.plot(x , tr_func_cold, 'b' , linewidth =2, label ='Cold')
plt.plot(x , tr_func_warm, 'r' , linewidth =2, label ='Warm')
plt.title('Triangular Fuzzy Set')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Value')
plt.legend()
plt.show()
```

```python
and_f = fuzz.fuzzy_and(x, tr_func_cold, x, tr_func_warm)
plt.plot(x , tr_func_cold, 'b' , linewidth =2, label ='Cold')
plt.plot(x , tr_func_warm, 'r' , linewidth =2, label ='Warm')
plt.plot(x , and_f[1], 'g' , linewidth =2, label ='And')
plt.title('Triangular Fuzzy Set')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Value')
plt.legend()
plt.show()
```

```python
or_f = fuzz.fuzzy_or(x, tr_func_cold, x, tr_func_warm)
plt.plot(x , tr_func_cold, 'b' , linewidth =2, label ='Cold')
plt.plot(x , tr_func_warm, 'r' , linewidth =2, label ='Warm')
plt.plot(x , or_f[1], 'g' , linewidth =2, label ='Or')
plt.title('Triangular Fuzzy Set')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Value')
plt.legend()
plt.show()
```

```python
not_f = fuzz.fuzzy_not(tr_func_cold)
plt.plot(x , tr_func_cold, 'b' , linewidth =2, label ='Cold')
plt.plot(x , not_f, 'g' , linewidth =2, label ='Not')
plt.title('Triangular Fuzzy Set')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Value')
plt.legend()
plt.show()
```
